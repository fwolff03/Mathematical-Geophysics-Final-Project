```python
from SWE_Solver import plotSWE
```

```python
# BASIC SETUP WITH DOCUMENATION 
Nx = 50 # Number of spatial grid points
B = lambda x: 1 # Bottom topography function. This function defines the topographic profile and should take spatial coordinates as input and return the bottom elevation at those coordinates.
h0 = [4.0 - B(_ / (Nx-1)) for _ in range(Nx)] #Initial water height profile. This should be an array of length Nx, representing the initial water height at different spatial locations.
u0 = [0.0 for _ in range(Nx)] #Initial water velocity profile. Similar to h0, this should be an array of length Nx, representing the initial water velocity at different spatial locations.
tEnd = 1.0 #End time of the simulation. The simulation starts at time t=0.
timePoints = [0.0, 0.5, 1.0] #List of time points at which you want to visualize the results.
steadyH = h0 
```

```python
# g (float, optional): Gravitational constant. Default is 1.
# method (str, optional): Method selection ('A', 'B' or 'C'). Default is 'C'.
# returns water depth and stream velocity
h, u = plotSWE(B, h0, u0, Nx, tEnd, timePoints, g=1, method='C')
```

```python
# cool example from theirs
Nx = 100
B = lambda x: 1
h0 = [2.0 for _ in range(Nx)]
def u(x):
    if 0 <= x <= 0.3:
        return 1
    if 0.3 < x < 0.7:
        return 0
    if 0.7 <= x <= 1:
        return -1
u0 = [u(_ / (Nx - 1)) for _ in range(Nx)]
tEnd = 0.5
timePoints = [0.0, .01, .02, .03, .04, ]
steadyH = None
```
