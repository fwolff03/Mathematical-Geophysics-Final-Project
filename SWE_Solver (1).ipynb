```python
# import plotSWE which is how they create visualizations of what is happening in the shallow water equations
from SWE_Solver import plotSWE
```

## Initial Simulation With Full Documentation

```python
# inital simulations conditions with full documentation
Nx = 50 # Number of spatial grid points
B = lambda x: -x # Bottom topography function. This function defines the topographic profile and should take spatial coordinates as input and return the bottom elevation at those coordinates.
h0 = [2 - B(_ / (Nx-1)) for _ in range(Nx)] #Initial water height profile. This should be an array of length Nx, representing the initial water height at different spatial locations.
u0 = [5 for _ in range(Nx)] #Initial water velocity profile. Similar to h0, this should be an array of length Nx, representing the initial water velocity at different spatial locations.
tEnd = 15 #End time of the simulation. The simulation starts at time t=0.
timePoints = [0.0, 7, 14, 15] #List of time points at which you want to visualize the results.
```

```python
# g (float, optional): Gravitational constant. Default is 1.
# method (str, optional): Method selection ('A', 'B' or 'C'). Default is 'C'.
# returns water depth and stream velocity at each time point
h, u = plotSWE(B, h0, u0, Nx, tEnd, timePoints, g=1, method='A')
```

## Simulation Doubbled h

```python
# conditions for doubbling h
# same Nx value as inital simulation
Nx = 50
# same B value as inital simulation
B = lambda x: -x
# h0 is now 4-B(_/(Nx-1)) rather than 2-B(_/(Nx-1))
h0 = [4 - B(_ / (Nx-1)) for _ in range(Nx)]
# u0 value is the same as the inital simulation
u0 = [5 for _ in range(Nx)]
# tEnd is the same as the inital simulation
tEnd = 15
# time point values are just t=15 or the steady state
timePoints = [15]
```

```python
# returns h and u for doubbled h
h, u = plotSWE(B, h0, u0, Nx, tEnd, timePoints, g=1, method='A')
```

## Simulation Doubbled u

```python
# conditions for doubbling u
# same Nx value as inital simulation
Nx = 50
# same B value as inital simulation
B = lambda x: -x
# h0 value is the same as the inital simulation
h0 = [2 - B(_ / (Nx-1)) for _ in range(Nx)]
# u0 value is now [10 for _ in range(Nx)] rather than [5 for _ in range(Nx)]
u0 = [10 for _ in range(Nx)]
# tEnd is the same as the inital simulation
tEnd = 15
# time point values are just t=15 or the steady state
timePoints = [15]
```

```python
# returns h and u for doubbled u
h, u = plotSWE(B, h0, u0, Nx, tEnd, timePoints, g=1, method='A')
```
